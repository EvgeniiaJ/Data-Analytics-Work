```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
data_frame = pd.read_csv('greenpointZillow.csv')
data_frame['json_data'] = data_frame['json_data'].astype(str)
data_frame.head()
```

```python
data_frame = data_frame[data_frame['json_data'] != 'nan']
data_frame.head()
```

```python
data_frame['json_data'].iloc[0]
```

```python
import json

def parse_json(s):
    result = {'Latitude': '', 'Longitude': '', 'error': ''}
    json_data = json.loads(s)
    
    try:
        result['Latitude'] = json_data.get('geo').get('latitude')
        result['Longitude'] = json_data.get('geo').get('longitude')
    except Exception as e:
        result['error'] = str(e)
    
    
    return result        
```

```python
series = data_frame['json_data'].apply(parse_json)
series
```

```python
output_data_frame = pd.DataFrame(index = data_frame.index)
output_data_frame['Latitude'] = series.apply(lambda d: "{}".format(d['Latitude'])).astype(float)
output_data_frame['Longitude'] = series.apply(lambda d: "{}".format(d['Longitude'])).astype(float)
```

```python
output_data_frame.head()
```

```python
column_map = {'ListingPrice': 'list-card-price',
             'BedroomCount': 'list-card-details',
             'BathroomCount': 'list-card-details 2',
             'RentSpaceSize': 'list-card-details 3',
             'TimeOnMarket': 'list-card-variable-text'}

for key, value in column_map.items():
    output_data_frame[key] = data_frame[value]
    if data_frame[value].empty:
        if key == 'BedroomCount':
            output_data_frame[key] = data_frame['list-card-label']
            print('1')
        elif key == 'BathroomCount':
            output_data_frame[key] = data_frame['list-card-label 2']
        elif key == 'RentSpaceSize':
            output_data_frame[key] = data_frame['ist-card-lebel 3']

output_data_frame.head()
```

```python
output_data_frame.to_csv("greenpoint_parsed_forrent.csv", index = False)
```

# Now we have the data parsed, let's display it on Google Map

```python
# find the average latitude and longitude
latitude = output_data_frame['Latitude'].astype(float).mean()
longitude = output_data_frame['Longitude'].astype(float).mean()

print('Average Latitude Value:', latitude)
print('Average Longitude Value:', longitude)
```

```python
choices = ["CartoDB dark_matter", "cartodb positron", "openstreetmap"]
folium_map = folium.Map(location = [latitude, longitude], zoom_start = 10, tiles = 'cartodb positron')
folium_map
```

### Adding an additional column

```python
additional_data_frame = output_data_frame
additional_data_frame['price'] = additional_data_frame['ListingPrice'].apply(lambda x: x.replace('$','').replace(',','').replace('/mo', '')).astype(float)
additional_data_frame['price_z_score'] = (additional_data_frame['price'] - additional_data_frame['price'].mean())/additional_data_frame['price'].std()
additional_data_frame['UsefulInfo'] = additional_data_frame.apply( lambda row: "Bedroom Count: {} bd/bds; Time on Market: {}".format(row['BedroomCount'], row['TimeOnMarket']), axis = 1)
additional_data_frame.head()
```
